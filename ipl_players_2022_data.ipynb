```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

response = requests.get("https://iplt20stats.com/ipl-2022-auction")
soup = BeautifulSoup(response.content, "lxml")

all_tables = soup.find_all("table", class_="table table-hover mb-0")
table_2 = all_tables[1]

# 1. Aik line mein headers nikal lein (PLAYER, AMOUNT, AUCTION INFO)
headers = [th.text.strip() for th in table_2.find_all("th")]

# 2. DataFrame banate waqt hi headers (columns) assign kar dein
df = pd.DataFrame(columns=headers)

# 3. Saari rows par loop chalayein 
# [1:] isliye lagaya kyunki heading (row 0) hum upar nikal chuke hain, ab sirf data chahiye
table_rows = table_2.find_all("tr")

for row in table_rows[1:]:
    # Data check karein (td)
    td_tags = row.find_all("td")
    if td_tags:
        row_data = [td.text.strip() for td in td_tags]
        
        # Nayi row insert karein
        length = len(df)
        df.loc[length] = row_data

# Display settings taake table khubsurat dikhe
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

# Final Result print karein
print(df)



# DataFrame ko CSV file mein save karne ke liye
df.to_csv("ipl_players_2022.csv", index=False)
```
